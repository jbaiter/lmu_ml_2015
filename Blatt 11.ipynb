### (b)

How do the distance change if it is also known that the first fifty words are contained in 750 of the total
N = 1000 documents in the set, while all other words only appear in 5 documents?

Wenn Informationen über die globalen Häufigkeiten der einzelnen Wörter bekannt sind, sollten diese mit den Dokumentvektoren verrechnet werden, üblich ist _idf_

$idf_j = log(\fra{N}{n_j})$

N ist die Anzahl aller Dokumente $n_j$ die Anzahl aller Dokumente, in denen Wort j vorkommt.

Hieraus wird _tfidf_ gebildet:

$tfidf_{i,j} = tf_{i,j}*idf_j$

In unserem Beispiel haben die ersten 50 Wörter eine $idf_j$ von $log(\frac{1000}{75}) = log(\frac{4}{3}) \approx 0.288$, die anderen Wörter ($j \in\{ 51, \dots, 100\}$) haben $idf_j = log(\frac{1000}{5}) = log(200) \approx 5.298$.

Wir erhalten also neue Merkmalsvektoren:

  | 1 | 2-10 | 11-50 | 51-60 | 61-100 
--|---|------|-------|-------|--------
A | $log(\frac{4}{3})$ | $log(\frac{4}{3})$ | $log(\frac{4}{3})$ | $\log(200)$ | $log(200)$ 
B | $log(\frac{4}{3})$ | $0$ | $0$ | $0$ | $0$ 
C | $log(\frac{4}{3})$ | $log(\frac{4}{3})$ | $log(\frac{4}{3})$ | $0$ | $0$ 
D | $0$ | $0$ | $2 * log(\frac{4}{3})$ | $2 * log(\frac{4}{3})$ | $0$
